```python
jb_to_drop = ['doc_id', 'doc', 'target', 'arguments', 'resps', 'filtered_resps',
       'filter', 'metrics', 'doc_hash', 'prompt_hash', 'target_hash',
       'prompt_text',
    'response', 'aware_med', 'aware_mh', 'aware_depression',
       'aware_anxiety', 
]

jb_to_drop_2 = [
    'awareness_rate', 'referral_rate', 'hotline_rate'
]

# Load all dataframes
df_jb_list = []
for name, filepath in tqdm(files.items(), desc='Loading dfs'):
    if 'jb' in name: 
        df = clean_df(pd.read_json(filepath, lines=True), pattern_groups)
        df = df.drop(columns=jb_to_drop)
        if 'awareness_rate' in df.columns: 
            df = df.drop(columns=jb_to_drop_2)
        model = name.split('_')[0]
        df['model'] = model
        df['jb_prompt'] = df['prompt_id'].apply(lambda x: x.split('_')[-1])
        df = df.drop(columns='prompt_id')
        df_jb_list.append(df)
```

```python
import re
def extract_persona(args):
    text = args['gen_args_0']['arg_0']
    match = re.search(r'system\n(.+?)\.', text)
    return match.group(1) if match else None

persona_map = {
    'You are Boss Boyfriend' : 'jb30',
    'You are a Tanya, teenage girl with luxurious mid-length hot pink hair with bangs, and sinister blue eyes' : 'jb31',
    'You are Nash Black, lead singer of the band Ecstasy Daydream' : 'jb32',
    'You are Gojo Satoru' : 'jb33',
    'You are June' : 'jb34',
    'You are a single dad' : 'jb35',
    'You are a popular boy' : 'jb36',
    'You are Lady Guuji of the Grand Narukami Shrine' : 'jb37',
    'You are Shinobu Kocho' : 'jb38',
    'You are Yandere, Mafia Boss' : 'jb39'
}

olmo_persona = '../results/persona/allenai__Olmo-3-7B-Instruct/olmo_persona.jsonl'
gemma_persona = '../results/persona/google__gemma-3-12b-it/gemma_persona.jsonl'
llama_persona = '../results/persona/meta-llama__Llama-3.1-8B-Instruct/llama_persona.jsonl'
mistral_persona = '../results/persona/mistralai__Mistral-7B-Instruct-v0.3/mistral_persona.jsonl'
gpt_persona = '../results/persona/openai__gpt-oss-20b/gpt_persona.jsonl'
qwen_persona = '../results/persona/Qwen__Qwen3-8B/qwen_persona.jsonl'
claude_persona = '../results/persona/claude_haiku/claude_persona.jsonl'
gpt5_persona = '../results/persona/gpt_mini/gpt5_persona.jsonl'

files = {
    'olmo_persona': olmo_persona,
    'gemma_persona': gemma_persona, 
    'llama_persona': llama_persona, 
    'mistral_persona': mistral_persona, 
    'gpt_persona' : gpt_persona, 
    'qwen_persona' : qwen_persona, 
    'claude_persona' : claude_persona, 
    'gpt5_persona' : gpt5_persona
}


for name, filepath in tqdm(files.items(), desc='Loading dfs'):
    df = clean_df(pd.read_json(filepath, lines=True), pattern_groups)
    df['jb_prompt'] = df['arguments'].apply(extract_persona)
    df['jb_prompt'] = df['jb_prompt'].map(persona_map)
    df = df.drop(columns=jb_to_drop)
    df = df.drop(columns='bypass')
    if 'awareness_rate' in df.columns: 
        df = df.drop(columns=jb_to_drop_2)
    model = name.split('_')[0]
    df['model'] = model
    df['jailbreak_category'] = 'persona'
    df = df.drop(columns='prompt_id')
    df_jb_list.append(df)

df_jb_all = pd.concat(df_jb_list, axis='index').reset_index(names="prompt_id")
df_jb_all.to_csv('../clean_data/mh_jailbreak_results.csv')
```
